Q1. What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?

A)
*Power BI, filters and slicers both refine data, but they differ in visibility and interactivity:*

 *a Basic Filter is a simple behind-the-scenes condition, an Advanced Filter allows complex criteria and a Slicer is a visual, interactive control placed directly on the report canvas for end users.*


#Basic Filter

  - Applied via the Filters pane.

  - Example: In your sales dataset, you could apply a basic filter to show only Category = Electronics.

  - Simple, but not interactive for end users.

#Advanced Filter

  - Allows conditions like Sales > 10,000 or Customer_Name contains “Customer_1”.

  - Example: You could filter your dataset to show only orders where Quantity ≥ 5 AND Region = South.

  - Useful for developers who want precise control.

#Slicer

  - A visual element (like a dropdown or slider) placed on the report canvas.

  - Example: Add a slicer for Region so users can toggle between North, South, East, West and see dashboards update instantly.

  - Types include List, Dropdown, Between (numeric/date ranges), Relative Date slicers.

  - Best for interactive dashboards where end users explore data themselves.

Q4. What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?

A)*A Visual-Level Filter in Power BI applies only to a single visual (like a chart, table, or card), while Page-Level Filters affect all visuals on one report page, and Report-Level Filters apply across the entire report. The key difference is scope: Visual filters are the most granular, Page filters cover one page, and Report filters are global.*

#**Examples**

**- Visual-Level Filter**

      - Applied directly to one visual in the Filters pane.

      - Example: In your Sales dataset, you could filter a Pie Chart to show only Product = Laptop. Other visuals remain unaffected.

**- Page-Level Filter**

      - Applied to the Filters pane at the page level.

      - Example: If you filter Region = South, every visual on that page (Pie Chart, Table, KPI cards) will show only South region data.

**- Report-Level Filter**

      - Applied to the Filters pane at the report level.

      - Example: If you filter Year = 2023, all visuals across all pages in the report will show only 2023 data.

Q6.What is the role of Drill-Down in Power BI? How does it improve data exploration?

A)
  $**-> Insert the Slicer**$

- In Power BI, go to the Visualizations pane.

- Select the Slicer visual (the filter icon).

- Drag Product Category into the Field well of the slicer.

- You’ll now see a list of product categories (e.g., Furniture, Electronics, Clothing).

$**-> Connect the Slicer to the Hierarchy Visual**$

- By default, slicers in Power BI filter all visuals on the page that share the same dataset.

- This means your Country → Region → Product Category hierarchy chart will automatically respond to the slicer.

$**-> Observe the Interaction**$

- Select a single Product Category (e.g., Furniture) in the slicer.

- The hierarchy chart now shows Sales by Country → Region → Furniture only.

- Select multiple categories (e.g., Furniture + Electronics).

- The hierarchy chart updates to show combined sales for those categories.

- Clear the slicer selection.

- The chart returns to showing all product categories.

$**-> Combine with Drill‑Down**$

- Start at Country level → drill into Region.

- With the slicer applied, you’ll only see regions filtered by the chosen product category.

- Drill further into Product Category → the slicer ensures only the selected categories appear.

 Q10. Explain the difference between Drill-Down and Drill-Through in Power BI using a real-world example

 A)
 # Drill‑Down

$**->Definition:**$

- Drill‑Down lets you explore hierarchical data within the same visual. You move from a high‑level summary to more detailed levels by clicking directly on chart elements.

    - Example (Retail Sales Dashboard):

    - You start with a bar chart showing Total Sales by Year.

    - Drill down → see Sales by Quarter.

    - Drill further → Sales by Month.

    - Finally → Sales by Day.

# Drill‑Through

$**->Definition:  **$

- Drill‑Through takes you to a different report page focused on a specific entity. It’s used when you want a detailed view of one item.

    - Example (Retail Sales Dashboard):

    - On your summary page, you see Sales by Product Category.

    - You right‑click on Laptop → choose Drill‑Through → Product Detail Page.

    - Power BI navigates to a new page showing Regionwise Laptop Sales, customer demographics, or profit margins.

    - A Back button lets you return to the summary page.

Q13. What is the “Keep All Filters” option in Drill-Through? How does it affect user interaction

A)*Keep All Filters” ensures that all slicer and filter selections from the summary page are carried over to the Drill‑Through detail page, making the interaction seamless and context‑aware. Without it, only the Drill‑Through field is passed, and other filters are reset*

#“Keep All Filters” Does

- Normally, when you Drill‑Through from a summary page to a detail page, only the Drill‑Through field (e.g., Product) is passed along.

- If you enable Keep All Filters, Power BI also carries over any other filters or slicer selections that were active on the summary page.

- This means the detail page respects the same filtering context you had before drilling through.

$**->Example**$

- Imagine you’re analyzing Sales by Product:

- On the summary page, you select Year = 2024 and Region = South using slicers.

- You right‑click on Laptop → Drill‑Through to the detail page.

- Without Keep All Filters:

- The detail page only filters by Laptop.

- Year and Region filters are ignored, so you see Laptop sales across all years and regions.

- With Keep All Filters:

- The detail page filters by Laptop + Year = 2024 + Region = South.

- You see Laptop sales specifically for South region in 2024 — exactly matching your context.
