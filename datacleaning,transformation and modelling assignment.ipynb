Q1)What is Data Loading in Power BI and why is it considered the first step of analysis?

ANS:-

Data loading is the process of importing data from various sources into Power BI for analysis and reporting.

Power BI allows:

-->Direct data loading

-->Data transformation before loading

It is considered the first step of analysis because:

Power BI cannot analyze data unless the data is imported.
It brings raw data into the system for cleaning, transforming, modeling, and visualization.
Accurate analysis depends on properly loaded data.

Example: Loading sales data from an Excel sheet into Power BI before creating reports and dashboards.



Q2)Explain the difference between “Load” and “Transform Data” in Power BI.

ANS:-
In Power BI, Load means importing data directly into Power BI without making any changes. It is used when the data is already clean and ready for analysis.

On the other hand, Transform Data opens the Power Query Editor, where users can clean and modify the data before loading it into Power BI. This includes removing null values, changing data types, filtering rows, renaming columns, merging tables, and other data preparation tasks.

In simple words, Load is used to import data as it is, while Transform Data is used to prepare and clean the data before analysis.

Q3) What is a Fact Table and a Dimension Table? Give examples from the dataset.

Fact Table

A Fact Table contains measurable values or numerical business data used for analysis.

Characteristics:

-->Contains metrics or calculations.

-->Usually large in size.

-->Connected to dimension tables using keys


Example:
A Sales Table containing:

-->Sales Amount

-->Quantity Sold

-->Profit

-->Order ID


-->Dimension Table

A Dimension Table contains descriptive information about the facts.

Characteristics:

-->Provides context to data

Contains attributes like names, categories,
locations, or dates

Example:

Customer Table → Customer Name, City
Product Table → Product Name, Category
Date Table → Month, Year

Relationship Example:

Fact Table: Sales
Dimension Tables: Customer, Product, Date

Q4)Why is Star Schema preferred over Snowflake Schema in Power BI?

ANS:-
A Star Schema is preferred in Microsoft Power BI because it is:

Simpler to understand
Faster for querying and reporting
Easier to maintain
Improves dashboard performance

In Star Schema:

One central fact table is connected directly to dimension tables.

Advantages over Snowflake Schema:

Fewer joins are required

Better performance

Easier data modeling

More user-friendly for report creation

Snowflake Schema

Dimension tables are further divided into sub-tables.
More complex and slower due to multiple joins.

Therefore, Power BI mainly recommends Star Schema for efficient analytics and visualization.
