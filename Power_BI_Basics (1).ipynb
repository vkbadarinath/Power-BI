Q1)  What is Power BI and why is it used in businesses?

A) *Power BI is Microsoft’s business intelligence and analytics platform that helps businesses turn raw data into actionable insights. It is widely used because it enables organizations to connect multiple data sources, build interactive dashboards, and make data-driven decisions quickly and effectively.*

# **What Power BI Is**

  - Business Intelligence Tool: A suite of applications (Power BI Desktop, Power BI Service, Power BI Mobile) for data visualization and reporting.

  - Data Integration: Connects to diverse sources like Excel, SQL databases, ERP/CRM systems, cloud services, and APIs.

  - Visualization & Reporting: Creates interactive dashboards, charts, and KPIs that are easy to interpret.

  - Collaboration: Reports can be shared across teams via the cloud, ensuring everyone works with the same insights

 # **Why Businesses Use Power BI**

 1. Improved Decision-Making

    - Provides real-time dashboards and KPIs.

    - Helps managers spot trends, risks, and opportunities quickly

 2. Data Unification

    - Combines scattered data from multiple systems (ERP, CRM, Excel, marketing tools) into a single view.

    - Reduces silos and ensures consistency.


  3. Cost-Effective & Scalable

      - Cloud-based service with flexible licensing.

      - Scales easily from small teams to enterprise-level deployments.

  4. Accessibility & Ease of Use

      - Designed for both analysts and non-technical users.

      - Features like drag-and-drop visuals and natural language queries make it user-friendly.

  5. Automation & AI Integration

      - Built-in AI features (Copilot, anomaly detection, predictive analytics).

      - Automates repetitive reporting tasks, saving time and reducing errors.

Q2) Name and explain the three main components of Power BI.

A) *The three main components of Power BI are Power BI Desktop, Power BI Service, and Power BI Mobile Apps. Together, they enable data preparation, cloud-based sharing, and mobile access, making Power BI a complete end-to-end business intelligence solution.*

#**Power BI Desktop**

- A free Windows application used to build reports and dashboards.

- Key Features:

  - Connects to multiple data sources (Excel, SQL Server, Oracle, cloud services).

  - Provides tools for data modeling, transformation, and visualization.

  - Includes Power Query (data cleaning), Power Pivot (data modeling with DAX), and Power View (visualization).

- Use Case: Analysts and developers use it to design reports offline before publishing them to the cloud.


#**Power BI Service**

- A cloud-based platform (also called Power BI Online).

- Key Features:

  - Allows users to publish, share, and collaborate on reports created in Power BI Desktop.

  - Supports real-time dashboards, scheduled data refreshes, and role-based access control.

  - Enables integration with Microsoft Teams, SharePoint, and other enterprise tools.

- Use Case: Business teams use it to access reports via browsers, ensuring everyone works with the same updated insights.

#**Power BI Mobile Apps**

- Mobile applications available for iOS, Android, and Windows devices.

- Key Features:

  - Provides on-the-go access to dashboards and reports.

  - Supports interactive filtering and drill-downs.

  - Push notifications for KPI changes or alerts.

- Use Case: Executives and field teams use it to monitor performance and make decisions anytime, anywhere.

Q3) Explain the Power BI workflow.

A) *Power BI’s workflow is the end-to-end process of turning raw data into interactive dashboards and business insights. It moves through stages of connecting to data, cleaning and transforming it, modeling relationships, creating reports, publishing to the cloud, and sharing across teams.*

#**Power BI Workflow Steps**

- Connect to Data Sources

- Power BI can connect to Excel, CSV, SQL databases, cloud services (Azure, Salesforce, Google Analytics), APIs, and SharePoint.

- Two modes:

  - Import Mode → Data is loaded into Power BI for faster performance.

  - DirectQuery Mode → Queries run directly on the source system in real time.

#**Data Cleaning & Transformation (Power Query)**

- Raw data is rarely ready for analysis.

- Common tasks:

  - Remove duplicates.

  - Handle missing values.

  - Change data types.

  - Rename columns.

  - Split/merge fields.

- Power Query records every step, making transformations repeatable and transparent.

#**Data Modeling**

- Define relationships between tables (e.g., Sales ↔ Customers).

- Choose correct keys (Primary/Foreign).

- Control filter flow across tables.

- A strong model ensures accurate totals, correct slicers, and reliable KPIs.

#**Calculations with DAX (Data Analysis Expressions)**

- Create calculated columns and measures.

- Examples:

  - Total Sales = SUM(Sales[Sales_Amount])

  - Profit Margin = DIVIDE([Profit], [Sales])

- DAX enables advanced analytics like year-over-year growth, running totals, and custom KPIs.

#**Report & Dashboard Creation**

- Build interactive visuals: charts, tables, maps, KPIs.

- Add slicers and filters for drill-down analysis.

- Design dashboards tailored to business needs (e.g., Sales Performance, Customer Insights).

#**Publish to Power BI Service**

- Upload reports from Power BI Desktop to the Power BI Service (cloud).

- Enables:

  - Collaboration across teams.

  - Scheduled refreshes.

  - Role-based access control.

  - Integration with Microsoft Teams and SharePoint.

#** Access via Power BI Mobile**

- Dashboards are available on iOS, Android, and Windows apps.

- Executives and field teams can monitor KPIs on the go.

- Push notifications alert users to changes in metrics.

Q4) List any four data cleaning tasks that can be performed in Power Query

A)
#Data Cleaning Tasks in Power Query

1.  Remove Duplicates

    - Identify and eliminate repeated rows to ensure unique records.

    - Example: Removing duplicate Order_ID entries in a sales dataset.

2. Handle Missing Values

    - Replace nulls with default values, averages, or flags.

    - Example: Filling missing Sales_Amount with the median or a placeholder like 0.

3. Change Data Types

    - Convert columns to the correct type (e.g., text → number, string → date).

    - Example: Converting "Three Thousand" into a numeric value 3000.

4. Standardize Formats

    - Ensure consistency in dates, text casing, or units.

    - Example: Converting mixed date formats (DD-MM-YYYY, YYYY/MM/DD) into ISO YYYY-MM-DD.

Q6) Define Data View, Report View, and Model View. Explain the purpose of each view.

A) IN Power BI (Data View, Report View, & Model View) these three main steps are plays a vitole role in power bi for analyzing the data in given report & provide a concrete analysis report for easily understanding everyone.

#Data View:

- Purpose: To inspect and clean the raw data after it’s loaded.

- What you see: A table-like grid showing rows and columns, similar to Excel.

- Why it matters:

  - Lets you check for missing values, duplicates, or incorrect data types.

  - Useful for validating transformations applied in Power Query.

  - Ensures the dataset is accurate before building visuals.


#Report View:

- Purpose: To design and interact with dashboards and visualizations.

- What you see: A canvas where you drag fields to create charts, tables, KPIs, and slicers.

- Why it matters:

  - This is where you build the actual reports that business users will consume.

  - Allows customization of layouts, themes, and interactivity.

  - Supports storytelling with data by combining visuals into a single dashboard.


#Model View:

- Purpose: To define relationships between tables and manage the data model.

- What you see: A diagram showing tables and their connections (like ER diagrams in databases).

- Why it matters:

  - Ensures correct relationships (one-to-many, many-to-many) between tables.

  - Controls how filters and slicers propagate across visuals.

  - Critical for accurate aggregations and avoiding misleading results.


    - *Data View → Think of it as your spreadsheet (raw data inspection)*.

    - *Report View → Your presentation slides (visual storytelling)*.

    - *Model View → Your blueprint (relationships and structure)*.

Q7)  Discuss the different data sources that Power BI supports.

A) *Power BI supports a wide range of data sources, including files, databases, cloud services, and online platforms. This flexibility allows businesses to connect virtually any dataset into their dashboards for analysis.*

#**Categories of Power BI Data Sources**

1. File-Based Sources

- Examples: Excel, CSV/Text, XML, JSON, PDF, SharePoint folders.

- Purpose: Common for importing structured or semi-structured data stored locally or in shared drives.

- Use Case: Loading daily sales reports from Excel or CSV exports.

2. Database Sources

- Examples: SQL Server, Oracle, MySQL, PostgreSQL, IBM DB2, Teradata, SAP HANA.

- Purpose: Direct connection to enterprise databases for transactional or analytical data.

- Use Case: Connecting to a company’s ERP or CRM database for live reporting.

3. Microsoft Azure Cloud Sources

- Examples: Azure SQL Database, Azure Synapse Analytics, Azure Analysis Services, Azure Blob Storage, Azure Data Lake.

- Purpose: Cloud-native data storage and analytics integration.

- Use Case: Accessing large-scale cloud data warehouses without downloading locally.

4. Online Services

- Examples: Salesforce, Google Analytics, Facebook, Dynamics 365, GitHub, MailChimp.

- Purpose: Pulls data directly from SaaS platforms and web APIs.

- Use Case: Marketing teams analyzing campaign performance from Google Analytics or MailChimp.

5. Other Sources

- Examples: Hadoop, Active Directory, Exchange, ODBC connections.

- Purpose: Extends connectivity to specialized or legacy systems.

- Use Case: Integrating HR data from Active Directory or logs from Hadoop clusters.

Q8) Split Owner Name to create two new columns as First Name and Last Name

A)
#Step-by-Step Instructions

1. Open Power Query

- In Power BI Desktop, after loading your dataset, click Transform Data.

- This opens the Power Query Editor.

2. Select the Column

 - Locate the column named Owner Name.

- Click on it to highlight.

3. Use Split Column Feature

- On the ribbon, go to Home → Split Column → By Delimiter.

- Choose the delimiter as Space (since names are separated by spaces).

- Select Split at the left-most delimiter (for First Name) and Split at the right-most delimiter (for Last Name).

4. Rename the Columns

- Power Query will create two new columns, typically named Owner Name.1 and Owner Name.2.

- Rename them to First Name and Last Name.

5. Apply Changes

- Click Close & Apply to save transformations and return to Power BI.

- Your dataset now has separate First Name and Last Name columns.
